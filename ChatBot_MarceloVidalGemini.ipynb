```python
from typing import TextIO
"""
At the command line, only need to run once to install the package via pip:

$ pip install google-generativeai
"""

import google.generativeai as genai
from google.colab import userdata
api_key = userdata.get('SECRET_KEY')
genai.configure(api_key=api_key)

# Set up the model
generation_config = {
  "temperature": 1,
}

safety_settings = [
  {
    "category": "HARM_CATEGORY_HARASSMENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_HATE_SPEECH",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
]

model = genai.GenerativeModel(model_name="gemini-1.5-pro-latest",
                              generation_config=generation_config,
                              safety_settings=safety_settings)
confirma="N"
while True:

 print(" *** ChatBot (Gemini - Google)para dica de presentes *** ")
 print(" ***         Projeto Imers√£o Alura / 2024            *** ")
 print(" ***         Autor (Marcelo Vidal) v.1.0             *** ")
 print(" ******************************************************* ")
 print(' Preencha os campos abaixo com os dados da pessoa a ser presenteada')
 print(' Observa√ß√£o : o preenchimento dos campos n√£o √© obrigat√≥rio, por√©m quanto mais dados forem informados melhor ser√° a resposta')
 print("     ");

 idade = input("1 - Idade da pessoa a ser presenteada : ")
 sexo = input("2 - Sexo por extenso : ")
 religiao = input("3 - Religi√£o : ")
 relacao = input("4 - Qual seu tipo de rela√ß√£o com essa pessoa (√© sua M√£e, Conjuge, Chefe, Amigo, Cunhado...) : ")
 datacomemorativa = input("5 - Se for para alguma data comemorativa descreva qual √© a data ( Anivers√°rio, casamento, dia das m√£es...) : ")
 hobby = input ("6 - O que a pessoa gosta de fazer / hobby (importante preencher esse campo) : ")
 trabalho = input("7 - Informa√ß√µes referente ao trabalho da pessoa : ")
 estilo = input("8 - Estilo da pessoa (extrovertida , t√≠mida, s√©ria.. ), pode ter v√°rios esilos : ")
 valorminimo = input("9 - Valor m√≠nimo para o presente (n√£o coloque o simbolo R$) (ex: 80) : ")
 valormaximo = input("10 - Valor m√°ximo do presente (n√£o coloque o simbolo R$) (ex:1000) : ")
 naoquer = input("11-  Descreva algo que a pessoa n√£o gosta e/ou que n√£o deve ser sugerido como presente. : ")
 informacao =  input("12 - Alguma outra informa√ß√£o que julgar importante (ex: tipo f√≠sico, etinia, o que a pessoa n√£o gosta..) : ")

 confirma=input("Confirma parametros informados ? Tecle(S=Sim , N=N√£o) : ")
 confirma=confirma.upper()
 if confirma!="S":
  continue
 if confirma=="S":
   break

print("Aguarde um momento estamos trabalhando .......")
texto_pesquisa=""

if len(relacao)>0 :
  texto_pesquisa = texto_pesquisa+" para um(a)  "+relacao

if len(idade)>0 :
  texto_pesquisa = texto_pesquisa+" com idade de "+idade+" anos"

if len(sexo)>0 :
  texto_pesquisa = texto_pesquisa+", do sexo "+sexo

if len(religiao)>0 :
  texto_pesquisa = texto_pesquisa+", da religi√£o  "+religiao

if len(datacomemorativa)>0 :
  texto_pesquisa = texto_pesquisa+" para comemorar o(a) "+datacomemorativa

if len(hobby)>0 :
  texto_pesquisa = texto_pesquisa+", e a pessoa gosta de ("+hobby+")"

if len(trabalho)>0 :
  texto_pesquisa = texto_pesquisa+", e trabalha com  ("+trabalho+")"

if len(estilo)>0 :
  texto_pesquisa = texto_pesquisa+", tem um estilo estilo ("+estilo+")."

if len(valorminimo)>0 :
  texto_pesquisa = texto_pesquisa+" O presente deve ter o valor maior que R$ "+valorminimo+"."

if len(valormaximo)>0 :
  texto_pesquisa = texto_pesquisa+" O presente deve ter o valor menor que R$ "+valormaximo+"."

if len(informacao)>0 :
  texto_pesquisa = texto_pesquisa+" Informa√ß√µes sobre o presente "+informacao+"."

if len(naoquer)>0 :
  texto_pesquisa = texto_pesquisa+" √â importante que o presente n√£o seja um desses a seguir ("+naoquer+")."

texto_pesquisa="Me de dicas de presente "+texto_pesquisa+" Quero imagens do presente e dicas de onde posso compra-lo."

convo = model.start_chat(history=[])

print("* Conte√∫do do request *")
print(texto_pesquisa)
print(" * ")

convo.send_message(texto_pesquisa)
print(convo.last.text)
```
